````python
import numpy as np
import pandas as pd
from helpers import stream_jsonl, write_jsonl
from datasets import load_dataset
````

## Load HumanEval Dataset

````python
dataset = load_dataset("openai/openai_humaneval")
counter = 0
for item in dataset["test"]:
    if counter == 1: break
    counter += 1
    for k,v in item.items():
        print(f'{k}:\n{v}\n')
    print('\n', '='*75, '\n', sep='')
````

````python
# convert to dict for convenience
dataset_dict = {i['task_id']:i for i in dataset['test']}
print(dataset_dict['HumanEval/0']['prompt'])
print(dataset_dict['HumanEval/0']['test'])
````

````python
file_name0 = 'temp/mistral_7b_complete_code_prompt_temperature1.0_topP1.0_completions_20250330_015510_2809.jsonl_results.jsonl'
file_name1 = 'temp/mistral_7b_complete_code_prompt_temperature1.0_topP1.0_completions_20250330_015510_2809_cleaned_partially.jsonl_results.jsonl'
file_name2 = 'temp/mistral_7b_complete_code_prompt_temperature1.0_topP1.0_completions_20250330_015510_2809_cleaned_fully.jsonl_results.jsonl'
````

````python
# load & convert results into dicts for convenience
results0 = list(stream_jsonl(file_name0))
results0_dict = {i['task_id']:i for i in results0}
print(len(results0_dict))

results1 = list(stream_jsonl(file_name1))
results1_dict = {i['task_id']:i for i in results1}
print(len(results1_dict))

results2 = list(stream_jsonl(file_name2))
results2_dict = {i['task_id']:i for i in results2}
print(len(results2_dict))
````

````python
idx = 0
for key in results1_dict:
    if not results1_dict[key]['passed'] and not results2_dict[key]['passed']:
        idx += 1        
        print(f'CASE {idx}. {key}')
        print('\nPROMPT:')
        print(dataset_dict[key]['prompt'])
        print('\n', '-'*75, '\n', sep='')
        print('\nRAW:')
        print('\nPassed:', results0_dict[key]['passed'])
        print('\nCompletion:\n', results0_dict[key]['completion'], sep='')
        print('\nResult:\n', results0_dict[key]['result'], sep='')
        print('\n', '-'*75, '\n', sep='')
        print('\nPARTIALLY CLEANED:')
        print('\nPassed:', results1_dict[key]['passed'])
        print('\nCompletion:\n', results1_dict[key]['completion'], sep='')
        print('\nResult:\n', results1_dict[key]['result'], sep='')
        print('\n', '-'*75, '\n', sep='')
        print('\nFULLY CLEANED:')
        print('\nPassed:', results2_dict[key]['passed'])
        print('\nCompletion:\n', results2_dict[key]['completion'], sep='')
        print('\nResult:\n', results2_dict[key]['result'], sep='')
        print('\n', '='*75, '\n', sep='')
        
print('Total cases:', idx)
````
